```python
import matplotlib.pyplot as plt
import seaborn as sn
party_redeem_sum = merged_2.groupby(['Name_of_the_Political_Party', 'Year_Purchase'])['Denominations_purchase'].sum().reset_index()

parties = party_redeem_sum['Name_of_the_Political_Party'].unique()


for party in parties:
    party_data = party_redeem_sum[party_redeem_sum['Name_of_the_Political_Party'] == party]

    plt.figure(figsize=(10, 6))
    sns.barplot(data=party_data, x='Year_Purchase', y='Denominations_purchase', color='skyblue')
    plt.title(f'Total Sum of Money Redeemed by {party} Each Year')
    plt.xlabel('Year')
    plt.ylabel('Total Denominations Redeemed')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent overlap of labels
    plt.show()
```
